```python
def leafon_clim_april(VAR, agg):
    
    files = glob(os.path.join(home, "Data", "Climate", VAR + "_" + "*.nc"))
    # order the files by year 
    names = [os.path.basename(x) for x in files]
    file_year = [int(x[-13:-9]) for x in names]
    files = [x for _, x in sorted(zip(file_year, files))]   
    K = -1 
    for file in files:
        print(file)
        K += 1
        ds = xr.open_mfdataset(file)[VAR]
        
        if K == 0:
            ds_april = ds.sel(time = np.isin(ds.time.dt.month, [4])).mean('time')
        else:
            ds_april = xr.concat([ds_april, ds.sel(time = np.isin(ds.time.dt.month, [4])).mean('time')], dim='year')
    
    ds_april.to_netcdf(path = os.path.join(home, "Data", "Climate", "Phenology", VAR + "_April.nc"))

def leafon_clim_march(VAR, agg):
    
    files = glob(os.path.join(home, "Data", "Climate", VAR + "_" + "*.nc"))
    # order the files by year 
    names = [os.path.basename(x) for x in files]
    file_year = [int(x[-13:-9]) for x in names]
    files = [x for _, x in sorted(zip(file_year, files))]   
    K = -1 
    for file in files:
        print(file)
        K += 1
        ds = xr.open_mfdataset(file)[VAR]
        
        if K == 0:
            ds_april = ds.sel(time = np.isin(ds.time.dt.month, [3])).mean('time')
        else:
            ds_april = xr.concat([ds_april, ds.sel(time = np.isin(ds.time.dt.month, [3])).mean('time')], dim='year')
    
    ds_april.to_netcdf(path = os.path.join(home, "Data", "Climate", "Phenology", VAR + "_March.nc"))


def leafon_clim_may(VAR, agg):
    
    files = glob(os.path.join(home, "Data", "Climate", VAR + "_" + "*.nc"))
    # order the files by year 
    names = [os.path.basename(x) for x in files]
    file_year = [int(x[-13:-9]) for x in names]
    files = [x for _, x in sorted(zip(file_year, files))]   
    K = -1 
    for file in files:
        print(file)
        K += 1
        ds = xr.open_mfdataset(file)[VAR]
        
        if K == 0:
            ds_april = ds.sel(time = np.isin(ds.time.dt.month, [5])).mean('time')
        else:
            ds_april = xr.concat([ds_april, ds.sel(time = np.isin(ds.time.dt.month, [5])).mean('time')], dim='year')
    
    ds_april.to_netcdf(path = os.path.join(home, "Data", "Climate", "Phenology", VAR + "_May.nc"))


def leafoff_clim_august(VAR, agg):
    
    files = glob(os.path.join(home, "Data", "Climate", VAR + "_" + "*.nc"))
    # order the files by year 
    names = [os.path.basename(x) for x in files]
    file_year = [int(x[-13:-9]) for x in names]
    files = [x for _, x in sorted(zip(file_year, files))]   
    K = -1 
    for file in files:
        print(file)
        K += 1
        ds = xr.open_mfdataset(file)[VAR]
        
        if K == 0:
            ds_april = ds.sel(time = np.isin(ds.time.dt.month, [8])).mean('time')
        else:
            ds_april = xr.concat([ds_april, ds.sel(time = np.isin(ds.time.dt.month, [8])).mean('time')], dim='year')
    
    ds_april.to_netcdf(path = os.path.join(home, "Data", "Climate", "Phenology", VAR + "_August.nc"))

def leafoff_clim_september(VAR, agg):
    
    files = glob(os.path.join(home, "Data", "Climate", VAR + "_" + "*.nc"))
    # order the files by year 
    names = [os.path.basename(x) for x in files]
    file_year = [int(x[-13:-9]) for x in names]
    files = [x for _, x in sorted(zip(file_year, files))]   
    K = -1 
    for file in files:
        print(file)
        K += 1
        ds = xr.open_mfdataset(file)[VAR]
        
        if K == 0:
            ds_april = ds.sel(time = np.isin(ds.time.dt.month, [9])).mean('time')
        else:
            ds_april = xr.concat([ds_april, ds.sel(time = np.isin(ds.time.dt.month, [9])).mean('time')], dim='year')
    
    ds_april.to_netcdf(path = os.path.join(home, "Data", "Climate", "Phenology", VAR + "_September.nc"))

def leafoff_clim_october(VAR, agg):
    
    files = glob(os.path.join(home, "Data", "Climate", VAR + "_" + "*.nc"))
    # order the files by year 
    names = [os.path.basename(x) for x in files]
    file_year = [int(x[-13:-9]) for x in names]
    files = [x for _, x in sorted(zip(file_year, files))]   
    K = -1 
    for file in files:
        print(file)
        K += 1
        ds = xr.open_mfdataset(file)[VAR]
        
        if K == 0:
            ds_april = ds.sel(time = np.isin(ds.time.dt.month, [10])).mean('time')
        else:
            ds_april = xr.concat([ds_april, ds.sel(time = np.isin(ds.time.dt.month, [10])).mean('time')], dim='year')
    
    ds_april.to_netcdf(path = os.path.join(home, "Data", "Climate", "Phenology", VAR + "_October.nc"))
```

```python
def gridmet_nc_to_tif(nc, template):
    
    # template 
    ds = gdal.Open(template)
    wkt = ds.GetProjection()
    trans = ds.GetGeoTransform()
    cols = ds.RasterXSize
    rows = ds.RasterYSize   
    
    mync = xr.open_dataset(nc)
    myarray = mync.to_array()
    myarray = myarray.to_numpy()
    if len(myarray.shape)==4:
        myarray = myarray[0,:,:,:]
    
    K = 1983
    for time in range(0, myarray.shape[0]):
        K +=1
        myslice = myarray[time,:,:]

        # create the output image
        if len(range(0, myarray.shape[0])) == 1:
            breakdown_path = nc.split("/")
            outname = breakdown_path[len(breakdown_path)-2] + "_" + breakdown_path[len(breakdown_path)-1][:-3] + ".tif"
            outpath = os.path.join(home, "Data", "Climate", 'Phenology',"tifs", outname)
        else:
            breakdown_path = nc.split("/")
            outname = breakdown_path[len(breakdown_path)-2] + "_" + breakdown_path[len(breakdown_path)-1][:-3] + "_" + str(K) + ".tif"
            outpath = os.path.join(home, "Data", "Climate", 'Phenology',"tifs", outname)

        driver = ds.GetDriver()
        outDs = driver.Create(outpath, cols, rows, 1, gdal.GDT_Float32)
        outBand = outDs.GetRasterBand(1)
        outBand.WriteArray(myslice)
        outDs.SetGeoTransform(trans)    
        srs = osr.SpatialReference()
        srs.ImportFromWkt(wkt)
        outDs.SetProjection(srs.ExportToWkt())

        outDs = None
```

```python
# make a raster template 
import rioxarray as rxr
template = "/Volumes/GoogleDrive/My Drive/Chapter2_mechanisms_forest_water_cycling/Data/Climate/template.tif"
all_nc = glob(os.path.join(home, "Data", "Climate", "Phenology", "*.nc"))
for nc in all_nc:
    gridmet_nc_to_tif(nc, template)
```
